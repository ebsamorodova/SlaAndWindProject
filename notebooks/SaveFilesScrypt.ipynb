```python
#!/usr/bin/env python3

import urllib.request
import re

URL = 'https://nomads.ncdc.noaa.gov/data/cfsr/'

x = urllib.request.urlopen(URL)
y = x.read().decode()
folders = re.findall(r'<a href="([\d]{6})/">', y)
for folder in folders[::-2]: # 200306
    print('Downloading', folder)
    filename = f'wnd10m.l.gdas.{folder}.grb2'
    urllib.request.urlretrieve(f'{URL}{folder}/{filename}', filename)
    if str(folder) == "200306": break
```
